```
def mutiple_metric(yt,yp):
  from statistics import mean
  from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
  acc=round(accuracy_score(yt,yp)* 100,2)
  preci=round(mean(precision_score(yt,yp, average=None))* 100,2)
  recall=round(mean(recall_score(yt,yp, average=None))* 100,2)
  f1=round(mean(f1_score(yt,yp, average=None))* 100,2)
  score={'test_accuracy':acc,'test_precision_macro':preci,'test_recall_macro':recall,'test_f1_micro':f1}
  return score
```

```
def nb_tree_svm(df,path):
  from sklearn.model_selection import train_test_split
  from sklearn.model_selection import KFold
  from sklearn.model_selection import GridSearchCV
  X=df.values
  y=X[:,-1]
  X=X[:,:-1]
  train, test = train_test_split(df, test_size=0.3, random_state=42, shuffle=True, stratify=y)
  X_train=train.values
  y_train=X_train[:,-1]
  X_train=X_train[:,:-1]
  X_test=test.values
  y_test=X_test[:,-1]
  X_test=X_test[:,:-1]
  #cv_inner = KFold(n_splits=10, shuffle=True, random_state=1)
  # Import Decision tree classifier
  from sklearn.tree import DecisionTreeClassifier
  # Create a Decision tree classifier model
  clf = DecisionTreeClassifier()
  # Hyperparameter Optimization
  parameters = {'max_features': ['log2', 'sqrt','auto'], 
              'criterion': ['entropy', 'gini'],
              'max_depth': [2, 3, 5, 10, 50], 
              'min_samples_split': [2, 3, 50, 100],
              'min_samples_leaf': [1, 5, 8, 10]
             }
  # Run the grid search
  grid_obj = GridSearchCV(clf, parameters,cv=10,n_jobs=-1)
  grid_obj = grid_obj.fit(X_train, y_train)

  # Set the clf to the best combination of parameters
  clf = grid_obj.best_estimator_

  # Train the model using the training sets 
  clf.fit(X_train, y_train)
  # Prediction on test set
  y_pred = clf.predict(X_test)
  tree_score=mutiple_metric(y_test,y_pred)
  add_result_to_csv(tree_score,df.shape[1]-1,path,'tree_hybrid')
  ###############
  # SVM Classifier
  # Creating scaled set to be used in model to improve the results
  from sklearn.preprocessing import StandardScaler
  sc = StandardScaler()
  X_train = sc.fit_transform(X_train)
  X_test = sc.transform(X_test)
  # Import Library of Support Vector Machine model
  from sklearn import svm

  # Create a Support Vector Classifier
  svc = svm.SVC()

  # Hyperparameter Optimization
  parameters = [
    {'C': [1, 10, 100, 1000], 'kernel': ['linear']},
    {'C': [1, 10, 100, 1000], 'gamma': [0.001, 0.0001], 'kernel': ['rbf']},
  ]

  # Run the grid search
  grid_obj = GridSearchCV(svc, parameters,cv=10,n_jobs=-1)
  grid_obj = grid_obj.fit(X_train, y_train)

  # Set the svc to the best combination of parameters
  svc = grid_obj.best_estimator_

  # Train the model using the training sets 
  svc.fit(X_train,y_train)
  # Prediction on test data
  y_pred = svc.predict(X_test)
  svc_score=mutiple_metric(y_test,y_pred)
  add_result_to_csv(svc_score,df.shape[1]-1,path,'svm_hybrid')
  ####################
  #Gaussian Naive Bayes

  # Import library of Gaussian Naive Bayes model
  from sklearn.naive_bayes import GaussianNB

  nb_classifier = GaussianNB()

  params_NB = {'var_smoothing': np.logspace(0,-9, num=100)}
  gs_NB = GridSearchCV(estimator=nb_classifier, 
                  param_grid=params_NB, 
                  cv=10,   # use any cross validation technique 
                 n_jobs=-1) 
  gs_NB.fit(X_train,y_train)
  # Prediction on test set
  y_pred = gs_NB.predict(X_test)
  nb_score=mutiple_metric(y_test,y_pred)
  add_result_to_csv(nb_score,df.shape[1]-1,path,'nb_hybrid')
  from sklearn.neighbors import KNeighborsClassifier
  from sklearn.model_selection import cross_val_score
  from sklearn.model_selection import cross_validate
  from sklearn.model_selection import StratifiedKFold
  from sklearn.neighbors import KNeighborsClassifier
  from sklearn.model_selection import LeaveOneOut
  scoring = ['accuracy', 'precision_macro','recall_macro','f1_micro']
  k_fold = LeaveOneOut()
  #######################################################################################################
  clf = KNeighborsClassifier(n_neighbors=1,p=2)
  knn_score=cross_validate(clf, X, y, cv=k_fold,scoring=scoring,n_jobs=-1, return_train_score=False)
  #fn=('%s/knn')%(results_csv)
  add_result_to_csv(knn_score,df.shape[1]-1,path,'knn_hybrid')
  ########################################################################################################
  #return nb_score,tree_score,svm_score,knn_score
```

```
def chart_writer(writer,sheet_name,xname,yname):
  # Access the XlsxWriter workbook and worksheet objects from the dataframe.
  workbook  = writer.book
  worksheet = writer.sheets[sheet_name]

  # Create a chart object.
  chart = workbook.add_chart({'type': 'column'})

  # Some alternative colors for the chart.
  colors = ['#ff0000','#ffa500', '#ffff00', '#008000', '#0000ff', '#4b0082','#ee82ee','#851e3e','#1e1f26','#ff3377','#c99789']

  # Configure the series of the chart from the dataframe data.
  for col_num in range(1, len(colors) + 1):
      chart.add_series({
          'name':       [sheet_name, 1, col_num],
          #'categories': [sheet_name, 1,1,1,8],
          'values':     [sheet_name, 1, col_num, len(colors)-1, col_num],
          'fill':       {'color':  colors[col_num - 1]},
          'overlap':    -10,
      })

  # Configure the chart axes.
  chart.set_x_axis({'name':xname})
  chart.set_y_axis({})

  chart.set_y_axis({'name': yname, 'major_gridlines': {'visible':True}})

  # Insert the chart into the worksheet.
  worksheet.insert_chart('C4', chart)

  # Close the Pandas Excel writer and output the Excel file.
  #writer.save()
```

```
## ttest
def sign_ttest(rvs1,rvs2):
  a1=stats.ttest_ind(rvs1,rvs2)
  #print(a1)
  thres=np.float64(0.05)
  #print(type(a1[1]))
  #print(type(thres))
  try:
    a11=round(a1[1],2)
  except:
    pass
  if a1[1]<thres:
    if np.mean(rvs1)<np.mean(rvs2):
      #print('+')
      return '+,'+str(a11)
    else:
      #print('-')
      return '-,'+str(a11)
  else:
    #print('==')
    return '=,'+str(a11)
```

```
def accstd(x, y):                      # this is a demo function that takes in two ints and 
    return str(x) +str(u"\u00B1")+str(y)
```

```
def accstdp(x, y):                      # this is a demo function that takes in two ints and 
    #p=p[0]
    y=str(y)
    try:
      return str(x) + str('(')+str(y.split(',')[0])+str(')')
    except:
      return str(x) 
```

```
def accstdwtl(x, y):                      # this is a demo function that takes in two ints and 
    y=str(y)
    try:
      return str(x) + str('(')+str(y.split(',')[1]) +str(')')
    except:
      return str(x) 
```
