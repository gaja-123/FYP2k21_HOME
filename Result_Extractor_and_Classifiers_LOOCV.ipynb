## Average

```
import os
import pandas as pd
# Getting the current work directory (cwd)
thisdir = '/content/drive/MyDrive/FYP/review1_1/Proposed'
parameter=['# of features selected','gbest']
ds_names=os.listdir(thisdir)
M=['M_1','M_2','M_3','M_4','M_5','M_6','M_8','M_10']
df_fs = pd.DataFrame(columns=M,index=ds_names)
df_acc = pd.DataFrame(columns=M,index=ds_names)
# r=root, d=directories, f = files
for r, d, f in os.walk(thisdir):
    for file in f:
        if file.endswith("knn_average.csv"):
            path=os.path.join(r, file)
            df=pd.read_csv(path)
            path=path.split("/")
            for i in parameter:
              val=df[df['Unnamed: 0']==i]['0']
              val=float(val)
              if i=='# of features selected':
                df_fs.loc[path[-4],path[-2]]=val
              if i=='gbest':
                df_acc.loc[path[-4],path[-2]]=val
df_fs.to_csv(thisdir+'averageFS.csv')
df_acc.to_csv(thisdir+'averageACC.csv')
```

## MAX

```
import os
import pandas as pd
# Getting the current work directory (cwd)
thisdir = '/content/drive/MyDrive/FYP/review1_1/Proposed'
parameter=['# of features selected','gbest']
ds_names=os.listdir(thisdir)
M=['M_1','M_2','M_3','M_4','M_5','M_6','M_8','M_10']
df_fs = pd.DataFrame(columns=M,index=ds_names)
df_acc = pd.DataFrame(columns=M,index=ds_names)
# r=root, d=directories, f = files
for r, d, f in os.walk(thisdir):
    for file in f:
        if file.endswith("knn.csv"):
            path=os.path.join(r, file)
            df=pd.read_csv(path)
            path=path.split("/")
            val=df.loc[df.shape[0]-1,'# of features selected']
            val=float(val)
            df_fs.loc[path[-4],path[-2]]=val
            val=df.loc[df.shape[0]-1,'gbest']
            val=float(val)
            df_acc.loc[path[-4],path[-2]]=val      
df_fs.to_csv(thisdir+'MAXFS.csv')
df_acc.to_csv(thisdir+'MAXACC.csv')
```

```
import pandas as pd
import os
import numpy as np
thisdir = '/content/drive/MyDrive/FYP/review1_1/Proposed'
ds_names=os.listdir(thisdir)
endwith_tuple=list()
for i in range(1,31):
  endwith_tuple.append(("iteration%d.csv")%(i))
endwith_tuple=tuple(endwith_tuple)
# r=root, d=directories, f = files
for r, d, f in os.walk(thisdir):
    for file in f:
        if file.endswith(endwith_tuple):
            path=os.path.join(r, file)
            df=pd.read_csv(path) 
            nb_tree_svm(df,path)
```

```
def nb_tree_svm(df,path):
  X=df.values
  y=X[:,-1]
  X=X[:,:-1]
  from sklearn import svm
  from sklearn.naive_bayes import GaussianNB
  from sklearn.neighbors import KNeighborsClassifier
  from sklearn.model_selection import cross_val_score
  from sklearn.model_selection import cross_validate
  from sklearn.model_selection import StratifiedKFold
  from sklearn.neighbors import KNeighborsClassifier
  from sklearn.model_selection import LeaveOneOut
  scoring = ['accuracy', 'precision_macro','recall_macro','f1_micro']
  k_fold = LeaveOneOut()
  clf = GaussianNB()
  nb_score=cross_validate(clf, X, y, cv=k_fold,scoring=scoring,n_jobs=-1, return_train_score=True)
  #fn=('%s/naive bayes')%(results_csv)
  add_result_to_csv(nb_score,df.shape[1]-1,path,'naive bayes_LOOCV')
  #########################################################################################################
  clf = svm.LinearSVC()
  svm_score=cross_validate(clf, X, y, cv=k_fold,scoring=scoring,n_jobs=-1, return_train_score=True)
  #fn=('%s/svm_linear')%(results_csv)
  add_result_to_csv(svm_score,df.shape[1]-1,path,'svm_linear_LOOCV')
  #######################################################################################################
  clf = KNeighborsClassifier(n_neighbors=1,p=2)
  knn_score=cross_validate(clf, X, y, cv=k_fold,scoring=scoring,n_jobs=-1, return_train_score=True)
  #fn=('%s/knn')%(results_csv)
  add_result_to_csv(knn_score,df.shape[1]-1,path,'knn_LOOCV')
  ########################################################################################################
  #return nb_score,tree_score,svm_score,knn_score
```

```
def add_result_to_csv(score,nsf,path,clfname):
  # importing the csv module
  import csv
  import re
  path1=path.split("/")
  c=re.findall(r'\d+',path1[len(path1)-1])[0]
  M=re.findall(r'\d+',path1[len(path1)-2])[0]
  ds_name=path1[len(path1)-4]
  results_csv=(('%s/%s/results_csv1')%(thisdir,ds_name))
  if not os.path.exists(results_csv):
    os.makedirs(results_csv)
  results_csv=("%s/%s")%(results_csv,path1[len(path1)-2])
  if not os.path.exists(results_csv):
    os.makedirs(results_csv)
  fn=("%s/%s")%(results_csv,clfname)
  print("File Name : ",fn)
  # making columns names of result file as below
  fieldnames = ['Iteration','M','# of features selected','Fit-time','Score-time','Train_Accuracy','Test_Accuracy','Train_precision','Test_precision','Train_recall','Test_recall','Train_F-measure','Test_F-measure']
  file_exists = os.path.isfile(('%s.csv')%(fn))
  # opening the result file in appending mode
  with open(('%s.csv')%(fn), 'a', newline='') as csvfile:
      writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
      # if file already exist then no write the column name else write the column names
      if not file_exists:
        writer.writeheader() 
      # get the mean of each filedname and store it 
      for i in score.keys():
        score[i]=np.mean(score[i])
      # writing the result row to the csv file
      writer.writerow({'Iteration':c, 'M':M,'# of features selected':nsf,'Fit-time':score['fit_time'],'Score-time':score['score_time'],'Train_Accuracy':score['train_accuracy'],'Test_Accuracy':score['test_accuracy'],'Train_precision':score['train_precision_macro'],'Test_precision':score['test_precision_macro'],'Train_recall':score['train_recall_macro'],'Test_recall':score['test_recall_macro'],'Train_F-measure':score['train_f1_micro'],'Test_F-measure':score['test_f1_micro']})
  return
```
